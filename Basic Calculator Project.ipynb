# Calculator Project in Python 

```python
print("---------------------------------------------------------\n")
print("************ WELCOME TO SIMPLE CALCULATOR ************** \n")
print("-----------------------------------------------------------")

def Addition(x,y):
    return(x+y)
def Substraction(x,y):
    return(x-y)
def Multiplication(x,y):
    return(x*y)
def Division(x,y):
    return(x/y)
def Module(x,y):
    return(x%y)

print("1.Addition")
print("2.Substraction")
print("3.Multiplication")
print("4.Division")
print("5.Module")
print("================================================")

while True:
    choice = input("Enter choice(1/2/3/4/5):")
    
    if choice in('1','2','3','4','5'):
        num1 = float(input("Enter first choice number:"))
        num2 = float(input("Enter Second  choice number:"))    
                  
    if choice == '1':
        print(num1,"+",num2,"=",Addition(num1,num2))
        print("------------------------------------------")
    if choice == '2':
        print(num1,"-",num2,"=",Substraction(num1,num2))
        print("------------------------------------------")
    if choice == '3':
        print(num1,"*",num2,"=",Multiplication(num1,num2))
        print("------------------------------------------")
    if choice == '4':
        print(num1,"/",num2,"=",Division(num1,num2))
        print("-------------------------------------------")
    if choice == '5':
        print(num1,"%",num2,"=",Module(num1,num2))
        print("------------------------------------------")
    
    next_calculation = input("Lets do next_calculaton(Yes/No):")
    if next_calculation =="No":
        print("=================================================")
        print("THANKS FOR USING SIMPLE CALCULATOR \n")
        print("HAVE A NICE DAY")
        break
    
else:
    print("Invalild Input")
    print("=================================================================")
    print("THANKS FOR USING SIMPLE CALCULATOR \n")
    print("HAVE A NICE DAY")
        
```
